```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


# 获取累计解释方差
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance_ratio)

# 找到解释 50%、80%、90% 方差所需的主成分数量
thresholds = [0.5, 0.8, 0.9]
threshold_indices = [np.argmax(cumulative_variance >= t) + 1 for t in thresholds]

# 画累计解释方差曲线
plt.figure(figsize=(8, 5))
plt.plot(np.arange(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='-')
plt.axhline(y=0.5, color='r', linestyle='--', label="50% Variance")
plt.axhline(y=0.8, color='g', linestyle='--', label="80% Variance")
plt.axhline(y=0.9, color='b', linestyle='--', label="90% Variance")

# 标记对应的主成分数量
for t, idx in zip(thresholds, threshold_indices):
    plt.scatter(idx, cumulative_variance[idx - 1], color='black', zorder=3)
    plt.text(idx, cumulative_variance[idx - 1] - 0.05, f"{round(100*idx/len(cumulative_variance),2)} % PCs", ha='center', fontsize=10, color='black')

plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Cumulative Explained Variance with Markers")
plt.legend()
plt.grid()
plt.show()
```

```python
import os
import torch
import numpy as np
from tqdm import tqdm
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
import os
import sys
sys.path.append('/cpfs04/user/hanyujin/rule-gen/rule_tokenizer')
from torchvision import transforms
from utils.data import random_crop_arr, center_crop_arr
import ruamel.yaml as yaml
from torchvision.transforms import Compose, Resize, ToTensor, Normalize
sys.path.append('/cpfs04/user/hanyujin/rule-gen/rule_tokenizer')
from utils.model import build_tokenizer
vae_config = "/cpfs04/user/hanyujin/rule-gen/rule_tokenizer/configs/in1k/exp007-aediff-16.yaml"
vae_ckpt = "/cpfs04/user/hanyujin/rule-gen/experiments/tokenizer/mirror-exp001-aediff-16/checkpoints/0040000.pt"
# vae_ckpt = "/cpfs04/user/hanyujin/rule-gen/experiments/tokenizer/wrsd-exp009-ae-16/checkpoints/0032000.pt"
vae, vae_model, vae_embed_dim, dit_input_size, latents_bias, latents_scale, vae_1d = build_tokenizer(vae_config, vae_ckpt)
if vae_config is not None:
    with open(vae_config, 'r', encoding='utf-8') as f:
        file_yaml = yaml.YAML()
        config_args = file_yaml.load(f)
# 数据路径
data_path = "/cpfs04/user/hanyujin/rule-gen/datasets/mirrors_train"
import torch
from torchvision import transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, Dataset
import random

# 定义数据增强
base_transform = transforms.Compose([
    transforms.Resize((64, 64)),  # 统一尺寸
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

augment_transform = transforms.Compose([
    transforms.RandomRotation(degrees=15),  # 随机旋转 -15°~15°
    transforms.RandomHorizontalFlip(p=0.5),  # 50% 概率水平翻转
    transforms.RandomResizedCrop(size=64, scale=(0.8, 1.0)),  # 随机裁剪后调整回 64x64
])

class AugmentedDataset(Dataset):
    def __init__(self, root, target_size=10000, transform=None, augment_transform=None):
        self.dataset = ImageFolder(root, transform=None)  # 先加载原始数据
        self.transform = transform  # 原始图片处理
        self.augment_transform = augment_transform  # 数据增强
        self.target_size = target_size
        self.original_size = len(self.dataset)
    
    def __len__(self):
        return self.target_size  # 返回目标扩增后的数据量
    
    def __getitem__(self, idx):
        # 通过 `idx` 计算对应原始数据
        original_idx = idx % self.original_size
        img, label = self.dataset[original_idx]  # 获取原始图片
        
        # 50% 机率增强数据，否则保持原图
        if random.random() > 0.5 and self.augment_transform:
            img = self.augment_transform(img)
        
        # 转换为张量
        if self.transform:
            img = self.transform(img)
        
        return img, label

# 创建增强数据集
data_path = "/cpfs04/user/hanyujin/rule-gen/datasets/WRSD"
augmented_dataset = AugmentedDataset(
    root=data_path, 
    target_size=10000,  # 目标扩增到 10,000 张
    transform=base_transform,
    augment_transform=augment_transform
)

# DataLoader
dataloader = DataLoader(augmented_dataset, batch_size=4, shuffle=False, num_workers=4)


# 确保 VAE 处于评估模式
vae.cuda().eval()

# 存储所有 latent features
all_latents = []

# 遍历整个数据集
for imgs, _ in tqdm(dataloader, desc="Extracting Latents"):
    imgs = imgs.cuda()  # 只对图片进行 CUDA 迁移
    with torch.no_grad():
        latent_sample_ext, _, _ = vae.encode(imgs)  # 编码
        all_latents.append(latent_sample_ext.cpu().numpy())  # 存入 list（先放到 CPU）

# 拼接所有 latent
all_latents = np.concatenate(all_latents, axis=0)  # [N, C, H, W] 或其他 shape

# **保存 latent 到 npz**
save_path = f"/cpfs04/user/hanyujin/rule-gen/rule_tokenizer/eval/latents/{config_args['dataset']}-{config_args['vq_model']}.npz"
os.makedirs(os.path.dirname(save_path), exist_ok=True)
np.savez(save_path, latents=all_latents)

print(f"Saved all latents to {save_path}, shape: {all_latents.shape}")
```
